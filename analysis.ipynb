```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

# Settings
plt.style.use("ggplot")
sns.set_theme(style="whitegrid")

# Better figure size
plt.rcParams["figure.figsize"] = (10,6)
```

```python
df = pd.read_csv("data/cleaned_readytoworkDataset.csv")

df.head()
```

```python
df.info()
```

```python
df.describe(include="object")
```

##Distribution of Happiness Score

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="Happiness Score",
    bins=25,
    kde=True,
    color='darkgreen'
)

plt.title("Distribution of Happiness Score")
plt.show()
plt.savefig("image/happiness_score_distribution.png")
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="GDP per Capita",
    bins=25,
    kde=True,
    color='darkblue'
)

plt.title("Distribution of GDP per Capita")
plt.show()
plt.savefig("image/gdp_per_capita_distribution.png")
```

```python
sns.boxplot(
    x=df["Happiness Score"]
)
plt.title("Boxplot of Happiness Score")
plt.show()
plt.savefig("image/happiness_score_boxplot.png")
```

```python
top10 = df.sort_values("Happiness Score", ascending=False).head(10)

top10
```

```python
sns.barplot(
    data=top10,
    x="Happiness Score",
    y="Country",
    color='r'
)

plt.title("Top 10 Happiest Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.show()
plt.savefig("image/top10_happiest_countries.png")
```

```python
bottom10 = df.nsmallest(10,"Happiness Score")

sns.barplot(
    data=bottom10,
    x="Happiness Score",
    y="Country"
)

plt.title("Bottom 10 Least Happy Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Country") 
plt.show()
plt.savefig("image/bottom10_least_happy_countries.png")
```

##Bivariate Analysis
###How happiness depens on GDP,social_support etc.

```python
sns.regplot(
    data=df,
    x="GDP per Capita",
    y="Happiness Score"
)

plt.title("Happiness Score vs GDP per Capita")
plt.show()
plt.savefig("image/happiness_score_vs_gdp_per_capita.png")
```

```python
sns.regplot(
    data=df,
    x="Social Support",
    y="Happiness Score"
)
plt.title("Happiness Score vs Social Support")
plt.show()  
plt.savefig("image/happiness_score_vs_social_support.png")
```

```python
sns.regplot(
    data=df,
    x="Freedom to make life choices",
    y="Happiness Score"
)
plt.title("Happiness Score vs Freedom to Make Life Choices")
plt.show()
plt.savefig("image/happiness_score_vs_freedom_to_make_life_choices.png")
```

```python
sns.regplot(
    data=df,
    x="Healthy Life Expectancy",
    y="Happiness Score",
    color='g'
)
plt.title("Happiness Score vs Healthy Life Expectancy")
plt.show()
```

```python
sns.regplot(
    data=df,
    x="Perceptions of Corruption",
    y="Happiness Score",
    color='r'
)
plt.title("Happiness Score vs Perceptions of Corruption")
plt.show()
plt.savefig("image/happiness_score_vs_perceptions_of_corruption.png")
```

##Correlation Analysis

```python
corr = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(12,8))

cm=sns.color_palette("rocket_r", as_cmap=True)
sns.heatmap(
    corr,
    annot=True,
    cmap=cm,
    fmt=".2f"
)

plt.title("Correlation Matrix")
plt.show()
plt.savefig("image/correlation_matrix.png")
```

```python
columns = [
    "Happiness Score",
    "GDP per Capita",
    "Social Support",
    "Healthy Life Expectancy",
    "Freedom to make life choices"
]

sns.pairplot(df[columns])
```

##Time-Series Analysis

```python
yearly = df.groupby("Year")["Happiness Score"].mean()

yearly.plot(marker="o")
plt.title("Average Happiness Score Over Years")
plt.xlabel("Year")
plt.ylabel("Average Happiness Score")
plt.show()
```

```python
corr["Happiness Score"].sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
corr["Happiness Score"].sort_values(ascending=False).plot(kind='barh', color='g')

plt.title("Correlation of Factors with Happiness Score")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Factors")
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
plt.savefig("image/correlation_with_happiness_score.png")
```

Outlier Detection

```python
import seaborn as sns
import matplotlib.pyplot as plt

# আপনার কলামগুলোর নাম
cols_to_plot = ["Happiness Score", "GDP per Capita", "Social Support", 
                "Freedom to make life choices", "Healthy Life Expectancy"]

plt.figure(figsize=(12, 6))


sns.boxplot(data=df[cols_to_plot])

plt.title("Outlier Detection for Happiness Factors")
plt.xticks(rotation=30) 
plt.show()
plt.savefig("image/outlier_detection_happiness_factors.png")
```
