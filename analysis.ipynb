##Distribution of Happiness Score

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="Happiness Score",
    bins=25,
    kde=True,
    color='darkgreen'
)

plt.title("Distribution of Happiness Score")
plt.show()
```
